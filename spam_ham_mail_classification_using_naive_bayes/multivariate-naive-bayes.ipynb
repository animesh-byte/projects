```python
samples=len(dataset[1])
x_train = dataset[0,0:int(0.7*samples)+1]
y_train = dataset[1,0:int(0.7*samples)+1]

x_test = dataset[0,int(0.7*samples)+1:samples+1]
y_test = dataset[1,int(0.7*samples)+1:samples+1]
```

```python
def create_dict(x_train,x_test,y_train,y_test):
    total_words=[]
    spam_words=[]
    ham_words=[]
    for i in range(len(x_train)):
        spamWordsTemp = []
        hamWordsTemp = []
        wordsTemp = []
        total_words=total_words+x_train[i].split(" ")
        if(y_train[i]=='spam'):
            spamWordsTemp+= x_train[i].split(" ")
        else:
            hamWordsTemp+= x_train[i].split(" ")
        total_words += list(set(wordsTemp))
        spam_words += list(set(spamWordsTemp))
        ham_words += list(set(hamWordsTemp))
    total_words_set = set(total_words)
    spam_words_set = set(spam_words)
    ham_words_set = set(ham_words)
    
    total_words_dict = {}
    spam_words_dict = {}
    ham_words_dict = {}
    for i in total_words:
        total_words_dict[i] = total_words.count(i)
    for i in spam_words:
        spam_words_dict[i] = spam_words.count(i)
    for i in ham_words:
        ham_words_dict[i] = ham_words.count(i)
    return total_words_dict,spam_words_dict,ham_words_dict,total_words,spam_words,ham_words
```

```python
def probabilityGivenSpam(w, dictSpam, total_spam):
    return (dictSpam[w])
```

```python
def probabilityGivenHam(w, dictHam, total_ham):
    return (dictHam[w])
```

```python
def probabilityWord(w, dictAll, total_words):
    return dictAll[w] / total_words
```

```python
def probabilitySpam(mess, dictSpam, spamWords, dictAll, words, spamCount):
    num = den = 1
    listSpam = list(set(mess.split()))
    for w in listSpam:
        if w in spamWords:
            num *= probabilityGivenSpam(w, dictSpam, len(spamWords)) / spamCount
    return num / den
```

```python
def probabilityHam(mess, dictHam, hamWords, dictAll, words, hamCount): 
    num = den = 1
    listHam = list(set(mess.split()))
    for w in listHam:
        if w in hamWords:
            num *= probabilityGivenHam(w, dictHam, len(hamWords)) / hamCount
    return num / den
```

```python
def predict(mess, dictAll, dictSpam, dictHam, words, spamWords, hamWords, spamCount, hamCount):
    if probabilitySpam(mess, dictSpam, spamWords, dictAll, words, spamCount) > probabilityHam(mess, dictHam, hamWords, dictAll, words, hamCount):
        return "ham"
    else:
        return "spam"
```

```python
def getaccuracy(X_test, Y_test, dictAll, dictSpam, dictHam, words, spamWords, hamWords, spamCount, hamCount):
    corr = 0
    for i in range(len(Y_test)):
        pred = predict(X_test[i], dictAll, dictSpam, dictHam, words, spamWords, hamWords, spamCount, hamCount)
        if Y_test[i] == pred:
            corr += 1
    acc = corr / len(Y_test) * 100
    return "Accuracy = " + str(acc)
```

```python
dictAll, dictSpam, dictHam, words, spamWords, hamWords = create_dict(x_train, y_train,y_train,y_test)
```

```python
spamCount=0
hamCount=0
for i in range(len(y_train)):
    if y_train[i] == "spam":
        spamCount += 1
    else:
        hamCount += 1
```

```python
print(getaccuracy(x_test, y_test, dictAll, dictSpam, dictHam, words, spamWords, hamWords, spamCount, hamCount))
```
