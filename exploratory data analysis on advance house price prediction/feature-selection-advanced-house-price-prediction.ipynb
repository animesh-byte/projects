```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
dataset = pd.read_csv('../input/processed-x-train/X_train.csv')
Test = pd.read_csv('../input/processedxtest/X_test.csv')
```

```python
y_train=dataset['SalePrice']
X_train=dataset.drop(['Id','SalePrice'],axis=1)
y_test=Test['SalePrice']
X_test = Test.drop(['Id','SalePrice'],axis=1)

y_train=np.array(y_train)
y_train=pd.DataFrame(y_train.reshape(y_train.shape[0],1))
y_test=np.array(y_test)
y_test=pd.DataFrame(y_test.reshape(y_test.shape[0],1))
print(type(y_test),type(X_test))
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
# ### Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

feature_sel_model= SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_features = X_train.columns[feature_sel_model.get_support()]
```

```python
print('total features - ',X_train.shape[1])
print('selected features - ',len(selected_features))
print('features with coefficients shrank to zero: ',np.sum(feature_sel_model.estimator_.coef_ == 0))
```

```python
selected_features
```

```python
X_train=X_train[selected_features]
X_test = X_test[selected_features]
```
